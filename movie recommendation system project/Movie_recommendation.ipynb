```python
import pandas as pd
import numpy as np
import pickle
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import ast
import warnings
warnings.filterwarnings("ignore")
```

```python
movies=pd.read_csv(r"E:\10.python\project\movie_recemadation\tmdb_5000_credits.csv\tmdb_5000_movies.csv")
credits=pd.read_csv(r"E:\10.python\project\movie_recemadation\tmdb_5000_credits.csv\tmdb_5000_credits.csv")
```

```python
df=pd.merge(movies,credits,on="title")
```

```python
df
```

```python
# df.info()
```

```python
df1=df[["movie_id","title","overview","genres","keywords","cast","crew","popularity"]]
```

```python
df1.head(2)
```

```python
df1.dropna(inplace=True)
```

```python
def converter(string):
    list1=[]
    for i in ast.literal_eval(string):
        name=i["name"]
        list1.append(name)
    return list1
```

```python
for col in ["genres","keywords"]:
    df1[col]=df1[col].apply(converter)
```

```python
# df1["cast"][1]
```

```python
def cast_converter(string):
    list1=[]
    count=0
    for i in ast.literal_eval(string):
        if count < 3:
            name=i["name"]
            list1.append(name)
            count+=1
    return list1
```

```python
df1["cast"]=df1["cast"].apply(cast_converter)
```

```python
df1
```

```python
def crew_converter(string):
    list1=[]
    for i in ast.literal_eval(string):
        if i["job"]=="Director":
            name=i["name"]
            list1.append(name)
    return list1
```

```python
df1["crew"]=df1["crew"].apply(crew_converter)
```

```python
df1
```

```python
def space_remove(input_list):
    list1=[]
    for i in input_list:
        result=i.replace(" ","")
        list1.append(result)
    return list1
        
```

```python
for col in ["cast","genres","keywords","crew"]:
    df1[col]=df1[col].apply(space_remove)
```

```python
df1["overview"]=df1["overview"].apply(lambda x:x.split())
```

```python
df1["tag"]=df1["overview"]+df1["cast"]+df1["genres"]+df1["keywords"]+df1["crew"]
```

```python
df1["tag"]=df1["tag"].apply(lambda x:" ".join(x))
```

```python
df1.iloc[0,8]
```

```python
df1.drop(["cast","genres","keywords","crew","overview"],axis=1,inplace=True)
```

```python
df1["tag"]=df1["tag"].apply(lambda x:x.lower())
```

```python
df1
```

```python
cv=CountVectorizer(max_features=5000,stop_words="english")
```

```python
cv.fit(df1["tag"])
vector=cv.transform(df1["tag"])
```

```python
vector=vector.toarray()
```

```python
vector
```

```python
sm=cosine_similarity(vector)
```

```python
movie_index=df1[df1["title"]=="Gandhi"].index[0]
```

```python
def Recommend_movie(movie):
    movie_index=df1[df1["title"]==movie].index[0]
    list_of_recomd=sorted(list(enumerate(sm[movie_index])),reverse=True,key=lambda x:x[1])
    for i in list_of_recomd[1:6]:
        print(df1.iloc[i[0]].title)
```
